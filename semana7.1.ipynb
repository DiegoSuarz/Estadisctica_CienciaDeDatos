# Prueba de hipótesis para 2 poblaciones

## Prueba de hipótesis para la media

**Descripción del Problema:**

Un instituto de investigación quiere comparar los niveles de colesterol de dos poblaciones diferentes, una que sigue una dieta mediterránea y otra que sigue una dieta occidental. Se toma una muestra aleatoria de individuos de cada población y se mide su nivel de colesterol en mg/dL. Se conocen las varianzas poblacionales de ambas poblaciones. Los datos recogidos son los siguientes:

- Dieta Mediterránea: muestra de 50 individuos con una media de 190 mg/dL y una varianza poblacional de 400.
- Dieta Occidental: muestra de 60 individuos con una media de 205 mg/dL y una varianza poblacional de 450.


**Pregunta de Investigación:**
¿Existe una diferencia significativa en los niveles de colesterol entre las dos poblaciones?

**Hipótesis:**
- Hipótesis nula (H0): No hay diferencia en los niveles de colesterol entre las dos poblaciones (μ1 = μ2).
- Hipótesis alternativa (H1): Hay una diferencia en los niveles de colesterol entre las dos poblaciones (μ1 ≠ μ2).

**Prueba Estadística:**
Se utilizará la prueba Z para dos medias con varianzas conocidas.

```python
import numpy as np
import scipy.stats as stats

# Datos
n1 = 50  # tamaño de la muestra de la dieta mediterránea
mean1 = 190  # media de la muestra de la dieta mediterránea
var1 = 400  # varianza poblacional de la dieta mediterránea

n2 = 60  # tamaño de la muestra de la dieta occidental
mean2 = 205  # media de la muestra de la dieta occidental
var2 = 450  # varianza poblacional de la dieta occidental

# Nivel de significancia
alpha = 0.05

# Cálculo del estadístico Z
z = (mean1 - mean2) / np.sqrt((var1 / n1) + (var2 / n2))

# Valor p
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Estadístico Z: {z}")
print(f"Valor p: {p_value}")

# Interpretación de resultados
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa en los niveles de colesterol entre las dos poblaciones.")
else:
    print("No rechazamos la hipótesis nula: No existe una diferencia significativa en los niveles de colesterol entre las dos poblaciones.")
```

**Descripción del Problema:**

Una empresa farmacéutica quiere comparar la efectividad de dos medicamentos distintos (Medicamento A y Medicamento B) para reducir la presión arterial. Se seleccionan al azar dos grupos de pacientes, cada uno recibe uno de los dos medicamentos. Después de un mes de tratamiento, se mide la reducción de la presión arterial en milímetros de mercurio (mmHg) para ambos grupos. Los datos recogidos son los siguientes:

Medicamento A: 10, 15, 14, 18, 20, 16, 15, 14, 17, 19
Medicamento B: 12, 14, 11, 15, 13, 16, 12, 14, 13, 15

**Pregunta de Investigación:**
¿Existe una diferencia significativa en la reducción de la presión arterial entre los pacientes que tomaron Medicamento A y los que tomaron Medicamento B?

Hipótesis:
- Hipótesis nula (H0): No hay diferencia en la reducción de la presión arterial entre los dos medicamentos (μA = μB).
- Hipótesis alternativa (H1): Hay una diferencia en la reducción de la presión arterial entre los dos medicamentos (μA ≠ μB).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos de reducción de presión arterial
medicamento_A = [10, 15, 14, 18, 20, 16, 15, 14, 17, 19]
medicamento_B = [12, 14, 11, 15, 13, 16, 12, 14, 13, 15]

# Realizar prueba t para muestras independientes
t_stat, p_value = stats.ttest_ind(medicamento_A, medicamento_B)

print(f"Estadístico t: {t_stat}")
print(f"Valor p: {p_value}")

# Interpretación de resultados
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa en la reducción de la presión arterial entre los dos medicamentos. (0.04086 < 0.05)")
else:
    print("No rechazamos la hipótesis nula: No existe una diferencia significativa en la reducción de la presión arterial entre los dos medicamentos.")
```

```python
# Gráficos
data = {
    'Medicamento': ['A']*len(medicamento_A) + ['B']*len(medicamento_B),
    'Reducción': medicamento_A + medicamento_B
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Medicamento', y='Reducción', data=df)
plt.title('Distribución de la Reducción de la Presión Arterial por Medicamento')
plt.xlabel('Medicamento')
plt.ylabel('Reducción de la Presión Arterial (mmHg)')
plt.show()
```

## Prueba de hipótesis para la proporción

**Descripción del Problema:**

Una cadena de tiendas de ropa quiere evaluar si existe una diferencia significativa en la proporción de clientes satisfechos entre dos de sus tiendas (Tienda A y Tienda B). Se realiza una encuesta a una muestra de clientes de cada tienda y se registra si los clientes están satisfechos o no. Los resultados de la encuesta son los siguientes:

- Tienda A: 120 clientes encuestados, 90 satisfechos
- Tienda B: 150 clientes encuestados, 105 satisfechos

**Pregunta de Investigación:**
¿Existe una diferencia significativa en la proporción de clientes satisfechos entre la Tienda A y la Tienda B?

Hipótesis:
- Hipótesis nula (H0): No hay diferencia en la proporción de clientes satisfechos entre las dos tiendas (pA = pB).
- Hipótesis alternativa (H1): Hay una diferencia en la proporción de clientes satisfechos entre las dos tiendas (pA ≠ pB).

```python
pip install statsmodels
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de la encuesta (muestra aleatoria)
satisfechos_A = 90
total_A = 120
satisfechos_B = 105
total_B = 150

# Proporciones
prop_A = satisfechos_A / total_A
prop_B = satisfechos_B / total_B

# Prueba de hipótesis de la proporción para 2 poblaciones
count = np.array([satisfechos_A, satisfechos_B]) #creamos un arreglo con los datos
nobs = np.array([total_A, total_B])
stat, p_value = sm.stats.proportions_ztest(count, nobs)

print(f"Estadístico Z: {stat}")
print(f"Valor p: {p_value}")

# Interpretación de resultados
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa en la proporción de clientes satisfechos entre las dos tiendas.")
else:
    print("No rechazamos la hipótesis nula: No existe una diferencia significativa en la proporción de clientes satisfechos entre las dos tiendas.")
```

```python
# Gráficos
data = {
    'Tienda': ['A', 'B'],
    'Proporción de Satisfechos': [prop_A, prop_B]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Proporción de Satisfechos', data=df)
plt.title('Proporción de Clientes Satisfechos por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Proporción de Clientes Satisfechos')
plt.ylim(0, 1)
plt.show()
```

## Otros casos

**NOTA**

Una prueba t de dos muestras investiga si las medias de dos muestras de datos independientes difieren entre sí. En una prueba de dos muestras, la hipótesis nula es que las medias de ambos grupos son iguales. A diferencia de la prueba de una muestra en la que probamos con un parámetro de población conocido, la prueba de dos muestras solo involucra medias muestrales. Puede realizar una prueba t de dos muestras pasando con la función stats.ttest_ind().

*Ejemplo (T-test muestras pareadas)*

Generemos una muestra de datos sobre la edad de los votantes para Wisconsin y probémosla.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
np.random.seed(6)
```

```python
#Poblaciones finitas
population_ages1 = stats.poisson.rvs(loc=18, mu=35, size=150000)
population_ages2 = stats.poisson.rvs(loc=18, mu=10, size=100000)
population_ages = np.concatenate((population_ages1, population_ages2)) #unir los 2 arrays

#Muestras
minnesota_ages1 = stats.poisson.rvs(loc=18, mu=30, size=30)
minnesota_ages2 = stats.poisson.rvs(loc=18, mu=10, size=20)
minnesota_ages = np.concatenate((minnesota_ages1, minnesota_ages2)) #unir los 2 arrays

print(population_ages.mean())
print(minnesota_ages.mean())
```

```python
np.random.seed(12)
wisconsin_ages1 = stats.poisson.rvs(loc=18, mu=33, size=30)
wisconsin_ages2 = stats.poisson.rvs(loc=18, mu=13, size=20)
wisconsin_ages = np.concatenate((wisconsin_ages1, wisconsin_ages2)) #unir los 2 arrays

print( wisconsin_ages.mean() )
```

```python
stats.ttest_ind(a= minnesota_ages, b= wisconsin_ages, equal_var=False)    # Assume samples have equal variance?
```

La prueba arroja un valor p de 0,0907, lo que significa que hay un 9% de posibilidades de que veamos datos de muestra tan separados si los dos grupos evaluados son en realidad idénticos. Si estuviéramos usando un nivel de confianza del 95% **no podríamos rechazar la hipótesis nula**, ya que el valor p es mayor que el nivel de significancia correspondiente del 5%.

## Muestras pareadas

```python
np.random.seed(11)

before= stats.norm.rvs(scale=30, loc=250, size=100)

after = before + stats.norm.rvs(scale=5, loc=-1.25, size=100)

weight_df = pd.DataFrame({"weight_before":before,
                          "weight_after":after,
                          "weight_change":after-before})

weight_df.describe()             # Check a summary of the data
```

El resumen muestra que los pacientes perdieron alrededor de 1,23 libras en promedio después del tratamiento. Realicemos una prueba t pareada para ver si esta diferencia es significativa con un nivel de confianza del 95%:

```python
stats.ttest_rel(a = before,  b = after) #puerba t student  a un nivel de significancia del 0.05 
```

En el gráfico anterior, las áreas rojas indican errores de tipo I, asumiendo que la hipótesis alternativa no es diferente de la nula para una prueba bilateral con un nivel de confianza del 95%.

El área azul representa errores de tipo II que ocurren cuando la hipótesis alternativa es diferente de la nula, como lo muestra la distribución de la derecha. # ANOVA 1 factor

**Descripción del Problema:**

Una empresa de tecnología está interesada en saber si diferentes programas de capacitación afectan el rendimiento de sus empleados. Para ello, se seleccionan tres programas de capacitación distintos: Programa A, Programa B y Programa C. Después de completar los programas, se mide el rendimiento de los empleados mediante una puntuación en una evaluación estandarizada.

**Pregunta de Investigación:**

¿Existe una diferencia significativa en el rendimiento de los empleados según el programa de capacitación al que asistieron?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos de las puntuaciones
programa_A = [78, 82, 85, 88, 92]
programa_B = [75, 80, 83, 85, 89]
programa_C = [90, 92, 93, 95, 98]

# Realizar ANOVA de un factor
f_value, p_value = stats.f_oneway(programa_A, programa_B, programa_C)

print(f"Estadístico F: {f_value}")
print(f"Valor p: {p_value}")

# Interpretación de resultados
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula: Existe una diferencia significativa entre los programas de capacitación.")
else:
    print("No rechazamos la hipótesis nula: No existe una diferencia significativa entre los programas de capacitación.")
```

```python
# Gráficos
data = {
    'Programa': ['A']*len(programa_A) + ['B']*len(programa_B) + ['C']*len(programa_C),
    'Puntuación': programa_A + programa_B + programa_C
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Programa', y='Puntuación', data=df)
plt.title('Distribución de Puntuaciones por Programa de Capacitación')
plt.xlabel('Programa de Capacitación')
plt.ylabel('Puntuación')
plt.show()
```

# ANOVA 2 factores

**Caso de Uso para ANOVA de Dos Factores con Gráficos**

Descripción del Problema:


Un investigador quiere estudiar el efecto combinado del tipo de fertilizante y el tipo de suelo en el crecimiento de las plantas. Para ello, se utilizan tres tipos de fertilizantes (Fertilizante A, Fertilizante B, Fertilizante C) y dos tipos de suelo (Suelo 1 y Suelo 2). Se mide el crecimiento de las plantas en centímetros después de un mes.


**Preguntas de Investigación:**

- ¿Existe una diferencia significativa en el crecimiento de las plantas debido al tipo de fertilizante utilizado?
- ¿Existe una diferencia significativa en el crecimiento de las plantas debido al tipo de suelo utilizado?
- ¿Existe una interacción significativa entre el tipo de fertilizante y el tipo de suelo en el crecimiento de las plantas?

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de las puntuaciones
data = {
    'fertilizante': ['A']*6 + ['B']*6 + ['C']*6,
    'suelo': ['1']*3 + ['2']*3 + ['1']*3 + ['2']*3 + ['1']*3 + ['2']*3,
    'crecimiento': [25, 27, 26, 30, 32, 31, 22, 24, 23, 28, 29, 30, 20, 21, 22, 27, 28, 29]
}

df = pd.DataFrame(data)

# Definir el modelo de ANOVA de dos factores
model = ols('crecimiento ~ C(fertilizante) * C(suelo)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

# Interpretación de resultados
alpha = 0.05
for factor, row in anova_table.iterrows():
    p_value = row['PR(>F)']
    if p_value < alpha:
        print(f"Rechazamos la hipótesis nula para {factor}: Existe una diferencia significativa.")
    else:
        print(f"No rechazamos la hipótesis nula para {factor}: No existe una diferencia significativa.")
```
