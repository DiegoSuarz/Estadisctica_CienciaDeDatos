## Variable Aleatoria Discreta

### Bernoulli

Supongamos que queremos modelar el resultado de lanzar una moneda sesgada, donde "Ã©xito" representa obtener cara y "fracaso" representa obtener cruz. Definimos una variable aleatoria X que toma el valor 1 en caso de Ã©xito y 0 en caso de fracaso.

DefiniciÃ³n de la variable aleatoria discreta de Bernoulli:
X=1 , si se obtiene cara (Ã©xito)
X=0 , si se obtiene cruz (fracaso)

```python
import numpy as np
import matplotlib.pyplot as plt

# ParÃ¡metros de la distribuciÃ³n de Bernoulli
p = 0.3  # Probabilidad de Ã©xito (obtener cara)
valores = [0, 1]  # Posibles valores de la variable aleatoria

# Calcular las probabilidades de cada valor
probabilidades = [1-p, p]

# Graficar la funciÃ³n de probabilidad
plt.bar(valores, probabilidades, color=['red', 'blue'])
plt.title('FunciÃ³n de Probabilidad de la DistribuciÃ³n de Bernoulli')
plt.xlabel('Valor')
plt.ylabel('Probabilidad')
plt.xticks(valores, ['Fracaso', 'Ãxito'])
plt.ylim(0, 1)  # Asegurar que el eje y va de 0 a 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Una variable aleatoria discreta con distribuciÃ³n de Bernoulli, donde la moneda estÃ¡ sesgada con una probabilidad de Ã©xito de 0.3 (obtener cara) y 0.7 de fracaso (obtener cruz). Luego, se muestra un grÃ¡fico de barras que ilustra la frecuencia de Ã©xitos y fracasos en la muestra generada

### Binomial

Supongamos que estamos interesados en modelar el nÃºmero de caras que se obtienen al lanzar una moneda sesgada 10 veces, donde la probabilidad de obtener cara en cada lanzamiento es p=0.6.

DefiniciÃ³n de la variable aleatoria discreta binomial:
X  es el nÃºmero de caras obtenidas en 10 lanzamientos.
X  puede tomar valores enteros de 0 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# ParÃ¡metros de la distribuciÃ³n binomial
p = 0.6  # Probabilidad de Ã©xito (obtener cara en un lanzamiento)
n = 10  # NÃºmero de lanzamientos

# Calcular la funciÃ³n de probabilidad de la distribuciÃ³n binomial
valores = np.arange(0, n+1)  # Valores posibles de la variable aleatoria
probabilidades = np.array([np.math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in valores])

# Graficar la funciÃ³n de probabilidad
plt.bar(valores, probabilidades, color='blue', alpha=0.7)
plt.title('FunciÃ³n de Probabilidad de la DistribuciÃ³n Binomial')
plt.xlabel('NÃºmero de Caras')
plt.ylabel('Probabilidad')
plt.xticks(valores)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

La funciÃ³n de probabilidad de una distribuciÃ³n binomial con n=10 ensayos (lanzamientos de una moneda) y p=0.6 como la probabilidad de Ã©xito (obtener cara en un lanzamiento). La grÃ¡fica muestra las probabilidades de obtener cada cantidad de caras (desde 0 hasta 10) en los 10 lanzamientos.El valor mas probable que salga es el numero 6

### DistribuciÃ³n binomial

```python
import matplotlib.pyplot as plt
import math
# Define parameters
N = 100  # Total count
p1 = 0.1  # Probability for first distribution
p2 = 0.5  # Probability for second distribution
k_values = range(int(N + 1))  # Range of possible values for k

# Calculate probabilities using binomial distribution formula
rv_probs_1 = [math.comb(N, k) * p1**k * (1-p1)**(N-k) for k in k_values]
rv_probs_2 = [math.comb(N, k) * p2**k * (1-p2)**(N-k) for k in k_values]

# Define colors for the bars
colors = ['skyblue', 'lightcoral']

# Create the bar plot
plt.figure(figsize=(12.5, 4))
plt.bar(k_values, rv_probs_1, color=colors[0], alpha=0.6, label=f"N: {N}, p: {p1:.1f}")
plt.bar(k_values, rv_probs_2, color=colors[1], alpha=0.6, label=f"N: {N}, p: {p2:.1f}")

# Customize the plot
plt.legend(loc="upper left")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.title("Probability Mass Distributions of Binomial Random Variables")
# plt.xticks(k_values)  # Set x-axis ticks to k values
plt.xlim(0, N + 0.5)  # Set x-axis limits
plt.grid(True, linestyle='--', linewidth=0.5)  # Add a grid

# Display the plot
plt.show()
```

Problema:

Supongamos que estamos interesados en modelar la probabilidad de que una imagen seleccionada al azar del conjunto de datos represente un dÃ­gito especÃ­fico (por ejemplo, el dÃ­gito '5'). Queremos calcular la probabilidad de obtener exactamente 'k' imÃ¡genes del dÃ­gito '5' en una muestra de 'n' imÃ¡genes seleccionadas aleatoriamente del conjunto de datos.

Paso a Paso:

Cargar el Conjunto de Datos: Utilizaremos el conjunto de datos load_digits de sklearn.
Contar las Ocurrencias: Contar cuÃ¡ntas imÃ¡genes corresponden al dÃ­gito '5'.
Estimar la Probabilidad: Estimar la probabilidad p de que una imagen seleccionada al azar sea un '5'.
Aplicar la DistribuciÃ³n Binomial: Calcular la probabilidad de obtener exactamente k imÃ¡genes del dÃ­gito '5' en una muestra de n imÃ¡genes usando la distribuciÃ³n binomial.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits #data set que tiene incoporada las imagenes de numeros
from scipy.stats import binom #para poder obtener la distribucion binomial

# Cargar el conjunto de datos
digits = load_digits()

# Contar cuÃ¡ntas imÃ¡genes corresponden al dÃ­gito '5'
count_digit_5 = np.sum(digits.target == 5)
total_images = len(digits.target)
prob_digit_5 = count_digit_5 / total_images

print(f"Total de imÃ¡genes: {total_images}")
print(f"Total de imÃ¡genes del dÃ­gito '5': {count_digit_5}")
print(f"Probabilidad estimada de que una imagen sea un '5': {prob_digit_5:.4f}")
```

```python
# Definir el tamaÃ±o de la muestra (n) y el nÃºmero de Ã©xitos deseados (k)
n = 100  # TamaÃ±o de la muestra
k = 10   # NÃºmero deseado de imÃ¡genes del dÃ­gito '5'

# Calcular la probabilidad usando la distribuciÃ³n binomial
probabilidad_k_exitos = binom.pmf(k, n, prob_digit_5)
print(f"Probabilidad de obtener exactamente {k} imÃ¡genes del dÃ­gito '5' en una muestra de {n} imÃ¡genes: {probabilidad_k_exitos:.4f}")

# Graficar la distribuciÃ³n binomial
x = np.arange(0, n+1)
binomial_pmf = binom.pmf(x, n, prob_digit_5)

plt.bar(x, binomial_pmf, color='blue')
plt.xlabel('NÃºmero de imÃ¡genes del dÃ­gito \'5\'')
plt.ylabel('Probabilidad')
plt.title('DistribuciÃ³n Binomial de ImÃ¡genes del DÃ­gito \'5\'')
plt.show()
```

ConslusiÃ³n:

Este ejemplo muestra cÃ³mo utilizar la distribuciÃ³n binomial para modelar la probabilidad de un evento especÃ­fico (en este caso, la apariciÃ³n del dÃ­gito '5') en un conjunto de datos. La distribuciÃ³n binomial es una herramienta poderosa para modelar situaciones en las que hay dos posibles resultados (Ã©xito o fracaso) y se repiten mÃºltiples ensayos independientes.

### DistribuciÃ³n Poisson

```python
#Se va a generar 2 distribuciones, lambda1, lambda 2

import matplotlib.pyplot as plt
import numpy as np



# Define parameters
lambda1 = 10  # Lambda value for first distribution
lambda2 = 0.5  # Lambda value for second distribution
k_values = range(150)  # Range of possible values for k

# Calculate probabilities using Poisson distribution formula
rv_probs_1 = [np.exp(-lambda1) * (lambda1**k) / np.math.factorial(k) for k in k_values]
rv_probs_2 = [np.exp(-lambda2) * (lambda2**k) / np.math.factorial(k) for k in k_values]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot PMF for first distribution
plt.bar(k_values, rv_probs_1, label=f"Î» = {lambda1}", color='skyblue', alpha=0.6)

# Plot PMF for second distribution
plt.bar(k_values, rv_probs_2, label=f"Î» = {lambda2}", color='lightcoral', alpha=0.6)

# Customize the plot
plt.legend(loc="upper left")
plt.xlabel("k")
plt.ylabel("P(X = k)")
plt.title("Probability Mass Functions of Poisson Distributions")
# plt.xticks(k_values)
plt.grid(True, linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```

Problema:

Supongamos que estamos estudiando la frecuencia con la que se encuentran flores Iris con una longitud de pÃ©talo superior a un cierto umbral en un Ã¡rea de jardÃ­n. Queremos modelar esta frecuencia utilizando la distribuciÃ³n de Poisson.

Paso a Paso

- Cargar el Conjunto de Datos: Utilizaremos el conjunto de datos load_iris de sklearn.
- Definir el Umbral: Establecer un umbral para la longitud del pÃ©talo.
- Contar las Ocurrencias: Contar cuÃ¡ntas flores tienen una longitud de pÃ©talo superior al umbral.
- Calcular la Tasa de Ocurrencia: Calcular la tasa promedio de ocurrencia.
- Aplicar la DistribuciÃ³n de Poisson: Calcular la probabilidad de observar exactamente k flores con una longitud de pÃ©talo superior al umbral en un Ã¡rea dada utilizando la distribuciÃ³n de Poisson.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import poisson

# Cargar el conjunto de datos
iris = load_iris()
petal_lengths = iris.data[:, 2]  # Longitud de los pÃ©talos

# Definir el umbral para la longitud del pÃ©talo
umbral = 5.0 #encontrar la occurencia de que se encuentre en peltalo mayor a 5 ccm

# Contar cuÃ¡ntas flores tienen una longitud de pÃ©talo superior al umbral
count_above_threshold = np.sum(petal_lengths > umbral)
total_flores = len(petal_lengths)
tasa_ocurrencia = count_above_threshold / total_flores

print(f"Total de flores: {total_flores}")
print(f"Flores con longitud de pÃ©talo > {umbral}: {count_above_threshold}")
print(f"Tasa promedio de ocurrencia: {tasa_ocurrencia:.4f}")
```

```python
# Definir el nÃºmero de Ã©xitos deseados (k)
k = 3  # NÃºmero deseado de flores con longitud de pÃ©talo > umbral, que tan probable sea ocurran que 3 flores tengan un petalo mayor a 5cm

# Calcular la probabilidad usando la distribuciÃ³n de Poisson
probabilidad_k_exitos = poisson.pmf(k, tasa_ocurrencia * total_flores)
print(f"Probabilidad de observar exactamente {k} flores con longitud de pÃ©talo > {umbral}: {probabilidad_k_exitos:.4f}")

# Graficar la distribuciÃ³n de Poisson
x = np.arange(0, 20)
poisson_pmf = poisson.pmf(x, tasa_ocurrencia * total_flores)

plt.bar(x, poisson_pmf, color='purple')
plt.xlabel('NÃºmero de flores con longitud de pÃ©talo > umbral')
plt.ylabel('Probabilidad')
plt.title('DistribuciÃ³n de Poisson para la longitud de pÃ©talo > umbral')
plt.show()
```

ConlusiÃ³n:

Este ejemplo muestra cÃ³mo utilizar la distribuciÃ³n de Poisson para modelar la frecuencia de ocurrencia de un evento especÃ­fico (en este caso, flores con una longitud de pÃ©talo superior a un umbral) en un conjunto de datos. La distribuciÃ³n de Poisson es Ãºtil para modelar situaciones en las que los eventos ocurren de manera independiente y con una tasa constante en un intervalo de tiempo o espacio.

### DistribuciÃ³n Exponencial

```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
rates1 = 0.2  # Rate (lambda) for first distribution
rates2 = 0.5  # Rate for second distribution
rates3 = 1  # Rate for third distribution

# Generate x-values
x_vals = np.linspace(0, 10, 150)  # Range of x-values

# Define PDF functions
pdf_1 = rates1 * np.exp(-rates1 * x_vals)
pdf_2 = rates2 * np.exp(-rates2 * x_vals)
pdf_3 = rates3 * np.exp(-rates3 * x_vals)

# Define colors
colors = ['royalblue', 'orange', 'purple']

# Create the plot
plt.figure(figsize=(12.5, 3))

# Plot PDF with line and shaded area
plt.plot(x_vals, pdf_1, label=rf"$\lambda = {rates1:.1f}$", color=colors[0])
plt.fill_between(x_vals, pdf_1, color=colors[0], alpha=0.33, label=rf"PDF 1")

plt.plot(x_vals, pdf_2, label=rf"$\lambda = {rates2:.1f}$", color=colors[1])
plt.fill_between(x_vals, pdf_2, color=colors[1], alpha=0.33, label=rf"PDF 2")

plt.plot(x_vals, pdf_3, label=rf"$\lambda = {rates3:.1f}$", color=colors[2])
plt.fill_between(x_vals, pdf_3, color=colors[2], alpha=0.33, label=rf"PDF 3")

# Customize the plot
plt.legend(loc="upper right")
plt.xlabel("$x$")
plt.ylabel("Probability Density")
plt.title("Probability Distributions of Three Exponential Random Variables")
plt.grid(True, linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```

Problema:

Supongamos que estamos interesados en modelar el tiempo entre la apariciÃ³n de flores Iris con una longitud de pÃ©talo superior a un cierto umbral en un jardÃ­n. Queremos utilizar la distribuciÃ³n exponencial para estimar la probabilidad de que el tiempo entre dos flores consecutivas con longitudes de pÃ©talo superiores al umbral sea menor o igual a un cierto valor.

Paso a Paso

- Cargar el Conjunto de Datos: Utilizaremos el conjunto de datos load_iris de sklearn.
- Definir el Umbral: Establecer un umbral para la longitud del pÃ©talo.
- Calcular la Tasa de Ocurrencia: Calcular la tasa promedio de ocurrencia de flores con longitudes de pÃ©talo superiores al umbral.
- Aplicar la DistribuciÃ³n Exponencial: Calcular la probabilidad de que el tiempo entre dos ocurrencias consecutivas sea menor o igual a un cierto valor usando la distribuciÃ³n exponencial.


```python
#distribucion Exponencial, se modela tiempos de ocurrencia
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import expon

# Cargar el conjunto de datos
iris = load_iris()
petal_lengths = iris.data[:, 2]  # Longitud de los pÃ©talos

# Definir el umbral para la longitud del pÃ©talo
umbral = 5.0 #dentro del area cuanta probabilidad de que aparescan petalos mayores a 5 cm

# Calcular la tasa de ocurrencia de flores con longitud de pÃ©talo superior al umbral
count_above_threshold = np.sum(petal_lengths > umbral)
total_flores = len(petal_lengths)
tasa_ocurrencia = count_above_threshold / total_flores

print(f"Total de flores: {total_flores}")
print(f"Flores con longitud de pÃ©talo > {umbral}: {count_above_threshold}")
print(f"Tasa promedio de ocurrencia (lambda): {tasa_ocurrencia:.4f}")
```

```python
# Definir el tiempo entre eventos deseado
tiempo_entre_eventos = 1 / tasa_ocurrencia

# Calcular la probabilidad de que el tiempo entre dos ocurrencias consecutivas sea menor o igual a un cierto valor
valor_tiempo = 0.5  # Ejemplo de tiempo en el que estamos interesados
probabilidad = expon.cdf(valor_tiempo, scale=tiempo_entre_eventos)

print(f"Probabilidad de que el tiempo entre dos flores con longitud de pÃ©talo > {umbral} sea menor o igual a {valor_tiempo} unidades de tiempo: {probabilidad:.4f}")

# Graficar la distribuciÃ³n exponencial
x = np.linspace(0, 5, 1000)
expon_cdf = expon.cdf(x, scale=tiempo_entre_eventos)

plt.plot(x, expon_cdf, 'b-', label='CDF de la distribuciÃ³n exponencial')
plt.axvline(valor_tiempo, color='r', linestyle='--', label=f'Valor tiempo = {valor_tiempo}')
plt.xlabel('Tiempo entre ocurrencias')
plt.ylabel('Probabilidad acumulada')
plt.title('DistribuciÃ³n Exponencial del Tiempo entre Ocurrencias')
plt.legend()
plt.show()
```

ConclusiÃ³n:

Este ejemplo muestra cÃ³mo utilizar la distribuciÃ³n exponencial para modelar el tiempo entre eventos (en este caso, flores con una longitud de pÃ©talo superior a un umbral) en un conjunto de datos. La distribuciÃ³n exponencial es Ãºtil para modelar situaciones en las que los eventos ocurren de manera continua e independiente con una tasa constante.

### DistribuciÃ³n Normal

```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
mu1 = -2  # Mean (loc) for first distribution
sigma1 = 1 / 0.7  # Standard deviation (scale) for first distribution
mu2 = 0  # Mean for second distribution
sigma2 = 1 / 1  # Standard deviation for second distribution
mu3 = 3  # Mean for third distribution
sigma3 = 1 / 2.8  # Standard deviation for third distribution

# Generate x-values
x_vals = np.linspace(-8, 7, 150)  # Range of x-values

# Define probability density functions (PDF)
pdf_1 = (1 / (sigma1 * np.sqrt(2 * np.pi))) * np.exp(-(x_vals - mu1)**2 / (2 * sigma1**2))
pdf_2 = (1 / (sigma2 * np.sqrt(2 * np.pi))) * np.exp(-(x_vals - mu2)**2 / (2 * sigma2**2))
pdf_3 = (1 / (sigma3 * np.sqrt(2 * np.pi))) * np.exp(-(x_vals - mu3)**2 / (2 * sigma3**2))

# Define colors
colors = ['royalblue', 'orange', 'purple']

# Create the plot
plt.figure(figsize=(12.5, 3))

# Plot PDF with line and shaded area
plt.plot(x_vals, pdf_1, label=rf"$\mu = {mu1}, \sigma = {sigma1:.1f}$", color=colors[0])
plt.fill_between(x_vals, pdf_1, color=colors[0], alpha=0.33, label=rf"PDF 1")

plt.plot(x_vals, pdf_2, label=rf"$\mu = {mu2}, \sigma = {sigma2:.1f}$", color=colors[1])
plt.fill_between(x_vals, pdf_2, color=colors[1], alpha=0.33, label=rf"PDF 2")

plt.plot(x_vals, pdf_3, label=rf"$\mu = {mu3}, \sigma = {sigma3:.1f}$", color=colors[2])
plt.fill_between(x_vals, pdf_3, color=colors[2], alpha=0.33, label=rf"PDF 3")

# Customize the plot
plt.legend(loc="upper right")
plt.xlabel("$x$")
plt.ylabel("Probability Density")
plt.title("Probability Distributions of Three Normal Random Variables")
plt.grid(True, linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```

Problema

Supongamos que estamos interesados en modelar el grado de alcohol medio de las un conjunto de vinos utilizando la distribuciÃ³n normal. Queremos estimar la probabilidad de que el contenido de alcohol estÃ© entre 12 y 14.

Pasos:

- Cargar el Conjunto de Datos: Utilizaremos el conjunto de datos load_wine de sklearn.
- Seleccionar una CaracterÃ­stica Continua: Seleccionaremos el contenido medio de los vinos.
- Ajustar una DistribuciÃ³n Normal: Calcularemos la media y la desviaciÃ³n estÃ¡ndar de los contenidos de los vinos.
- Aplicar la DistribuciÃ³n Normal: Calcular la probabilidad de que el contenido medio de alcohol de los vinos caiga dentro de un cierto rango utilizando la distribuciÃ³n normal.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine #cargar datos de vinos
from scipy.stats import norm

# Cargar el conjunto de datos
wine = load_wine()
alcohol_content = wine.data[:, 0]  # Contenido de alcohol

# Calcular la media y la desviaciÃ³n estÃ¡ndar del contenido de alcohol
mean_alcohol = np.mean(alcohol_content)
std_alcohol = np.std(alcohol_content)
#std_alcohol = 2 #conociento empirico de otros experimentos

print(f"Media del contenido de alcohol: {mean_alcohol:.2f}")
print(f"DesviaciÃ³n estÃ¡ndar del contenido de alcohol: {std_alcohol:.2f}")
```

```python
# Definir el rango de contenido de alcohol de interÃ©s
lower_bound = 12  # LÃ­mite inferior
upper_bound = 14  # LÃ­mite superior

# Calcular la probabilidad usando la distribuciÃ³n normal
probabilidad = norm.cdf(upper_bound, mean_alcohol, std_alcohol) - norm.cdf(lower_bound, mean_alcohol, std_alcohol)
print(f"Probabilidad de que el contenido de alcohol estÃ© entre {lower_bound} y {upper_bound}: {probabilidad:.4f}")

# Graficar la distribuciÃ³n normal
x = np.linspace(mean_alcohol - 3*std_alcohol, mean_alcohol + 3*std_alcohol, 1000)
pdf = norm.pdf(x, mean_alcohol, std_alcohol)

plt.plot(x, pdf, 'b-', label='PDF de la distribuciÃ³n normal')
plt.axvline(lower_bound, color='r', linestyle='--', label=f'LÃ­mite inferior = {lower_bound}')
plt.axvline(upper_bound, color='g', linestyle='--', label=f'LÃ­mite superior = {upper_bound}')
plt.fill_between(x, 0, pdf, where=(x >= lower_bound) & (x <= upper_bound), color='gray', alpha=0.5)
plt.xlabel('Contenido de alcohol (%)')
plt.ylabel('Densidad de probabilidad')
plt.title('DistribuciÃ³n Normal del Contenido de Alcohol en el Vino')
plt.legend()
plt.show()
```

### DistribuciÃ³n normal estÃ¡ndar

puede transformar una desviacion normal, siempre que la media sea 0 y desviacion standart sea 1, se conoce como transformacion normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos de una distribuciÃ³n normal con media 5 y desviaciÃ³n estÃ¡ndar 2
mu = 5
sigma = 2
data = np.random.normal(mu, sigma, 1000)

# Estandarizar los datos
data_standardized = (data - mu) / sigma

# Graficar la distribuciÃ³n original y la estandarizada
plt.figure(figsize=(14, 6))

# Histograma de la distribuciÃ³n original
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)
plt.title('DistribuciÃ³n Normal Original (Î¼=5, Ï=2)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')

# Histograma de la distribuciÃ³n estandarizada
plt.subplot(1, 2, 2)
plt.hist(data_standardized, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
plt.title('DistribuciÃ³n Normal Estandarizada (Î¼=0, Ï=1)')
plt.xlabel('Valor EstÃ¡ndar')
plt.ylabel('Densidad de Probabilidad')

plt.suptitle('ConversiÃ³n de DistribuciÃ³n Normal a DistribuciÃ³n Normal EstÃ¡ndar')
plt.show()
```

### Teorema de lÃ­mite central

Se trata de que cuanta mas poblacion o muestras tenga, mas probable es que la distribucion se asemeje a una normal standart

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parÃ¡metros
num_samples = 1000  # NÃºmero de muestras
sample_size = 30    # TamaÃ±o de cada muestra

# Generar datos de una distribuciÃ³n uniforme
data = np.random.uniform(low=0, high=1, size=(num_samples, sample_size))

# Calcular la media de cada muestra
sample_means = np.mean(data, axis=1)

# Generar datos de una distribuciÃ³n uniforme con tamaÃ±o mayor
sample_size_large = 100
data_large = np.random.uniform(low=0, high=1, size=(num_samples, sample_size_large))
sample_means_large = np.mean(data_large, axis=1)

# Graficar histogramas de la media de las muestras
plt.figure(figsize=(14, 6))

# Histograma de medias de muestras pequeÃ±as
plt.subplot(1, 2, 1)
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title(f'Medias Muestrales (TamaÃ±o de muestra = {sample_size})')
plt.xlabel('Media Muestral')
plt.ylabel('Frecuencia')

# Histograma de medias de muestras grandes
plt.subplot(1, 2, 2)
plt.hist(sample_means_large, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title(f'Medias Muestrales (TamaÃ±o de muestra = {sample_size_large})')
plt.xlabel('Media Muestral')
plt.ylabel('Frecuencia')

plt.suptitle('IlustraciÃ³n del Teorema del LÃ­mite Central')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm

# Definir parÃ¡metros
lambda_param = 3   # ParÃ¡metro lambda para la distribuciÃ³n de Poisson
num_samples = 1000 # NÃºmero de muestras
sample_size_small = 30   # TamaÃ±o de cada muestra (pequeÃ±o)
sample_size_large = 100  # TamaÃ±o de cada muestra (grande)

# Generar datos de una distribuciÃ³n de Poisson
data_small = np.random.poisson(lam=lambda_param, size=(num_samples, sample_size_small))
data_large = np.random.poisson(lam=lambda_param, size=(num_samples, sample_size_large))

# Calcular la media de cada muestra
sample_means_small = np.mean(data_small, axis=1)
sample_means_large = np.mean(data_large, axis=1)

# Graficar histogramas de la media de las muestras
plt.figure(figsize=(14, 6))

# Histograma de medias de muestras pequeÃ±as
plt.subplot(1, 2, 1)
plt.hist(sample_means_small, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(sample_means_small), np.std(sample_means_small))
plt.plot(x, p, 'k', linewidth=2)
plt.title(f'Medias Muestrales (TamaÃ±o de muestra = {sample_size_small})')
plt.xlabel('Media Muestral')
plt.ylabel('Frecuencia')

# Histograma de medias de muestras grandes
plt.subplot(1, 2, 2)
plt.hist(sample_means_large, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(sample_means_large), np.std(sample_means_large))
plt.plot(x, p, 'k', linewidth=2)
plt.title(f'Medias Muestrales (TamaÃ±o de muestra = {sample_size_large})')
plt.xlabel('Media Muestral')
plt.ylabel('Frecuencia')

plt.suptitle('IlustraciÃ³n del Teorema del LÃ­mite Central usando DistribuciÃ³n de Poisson')
plt.show()
```

### SimulaciÃ³n de Monte Carlo

Problema:

En esta simulaciÃ³n, utilizaremos el mÃ©todo de Monte Carlo para estimar el valor de Pi. El mÃ©todo consiste en generar un gran nÃºmero de puntos aleatorios dentro de un cuadrado y contar cuÃ¡ntos de esos puntos caen dentro de un cÃ­rculo inscrito en el cuadrado. Luego, utilizamos la proporciÃ³n de puntos dentro del cÃ­rculo con respecto al total de puntos para calcular una estimaciÃ³n de Pi.

```python
import numpy as np
import matplotlib.pyplot as plt

# NÃºmero de puntos a generar
num_points = 200000

# Generar coordenadas aleatorias dentro del cuadrado unitario [-1, 1] x [-1, 1]
x = np.random.uniform(-1, 1, num_points)
y = np.random.uniform(-1, 1, num_points)

# Calcular la distancia desde el origen para cada punto
distance = np.sqrt(x**2 + y**2)

# Contar cuÃ¡ntos puntos estÃ¡n dentro del cÃ­rculo (distancia <= 1)
num_inside_circle = np.sum(distance <= 1)

# Estimar el valor de Pi usando la proporciÃ³n de puntos dentro del cÃ­rculo
estimated_pi = 4 * num_inside_circle / num_points

print(f"Valor estimado de Pi: {estimated_pi:.6f}")

# Visualizar la simulaciÃ³n
plt.figure(figsize=(8, 8))
plt.scatter(x, y, c=distance <= 1, cmap='coolwarm', marker='o', edgecolors='k', alpha=0.5)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('SimulaciÃ³n de Monte Carlo para Estimar Pi')
plt.axis('equal')
plt.show()
```
