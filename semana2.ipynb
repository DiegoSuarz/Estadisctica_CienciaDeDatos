>REVISANDO LOS ESTADÃSTICOS

```python
import pandas as pd
import numpy as np
import scipy.stats as sp #modulo para estadistica
import seaborn as sns

def encontrar_outliers(data,col):
    outliers=[]
    Q1=data[col].quantile(0.25) #quartil 1
    Q3=data[col].quantile(0.75) #quartil 3
    IQR=Q3-Q1 #rango intercuartilico
    lower_limit=Q1-(1.5*IQR)
    upper_limit=Q3+(1.5*IQR)
    for out1 in data[col]:
        if out1>upper_limit or out1<lower_limit:
            outliers.append(out1)
    return np.array(outliers) #retornar valores considerador outliers

def graficos(data,col):
    fix,(ax1,ax2)=plt.subplots(1,2,figsize=(15,5)) #dividieno sub areas de una grafica 1 fila, 2 columnas
    sns.boxplot(y=data[col],ax=ax1) #graficar un boxplot
    ax1.set_ylabel=col
    ax1.set_title('GrÃ¡fico de cajas de {}'.format(col))
    sns.histplot(data,x=col,kde=True,ax=ax2) #graficar un histograma
    ax2.set_title('Distribution plot of {}'.format(col))
    plt.axvline(data[col].mean(),color='red',linestyle='--')
    plt.axvline(data[col].median(),color='green',linestyle='--')
    plt.legend(['densidad','media','median','distribuciÃ³n'])

def analisis_numerico(data,col):
    print('Cantidad de outlier:',len(encontrar_outliers(data,col)))
    print('Media:',round(data[col].mean(),3))
    print('Mediana:',data[col].median())
    print('Q1:',data[col].quantile(0.25))
    print('Q3:',data[col].quantile(0.75))
    print('max:',data[col].max())
    print('AsimetrÃ­a:',round(sp.skew(data[col]),3))
    print('Kurtosis:',round(sp.kurtosis(data[col]),3))
    graficos(data,col)
```

Calcular la Curtosis:



```python
# Otra forma de calcular la curtosis

import numpy as np

# Datos de ejemplo
datos= df['Datos']
#Calcular la media y la desviaciÃ³n estÃ¡ndar
media = np.mean(datos)
desviacion_estandar = np.std(datos)

#Calcular las diferencias al cuadrado entre cada punto de datos y la media
diferencias_cuarta_potencia = (datos - media) ** 4

#Calcular el cuarto momento central
cuarto_momento = np.mean(diferencias_cuarta_potencia)

#Calcular la curtosis
curtosis = cuarto_momento / (desviacion_estandar ** 4)

print("Curtosis:", curtosis-3)
```

>DESVIACIÃN MEDIA



```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

def desviacion_media(datos):
    n = len(datos)
    media = sum(datos) / n
    #media = np.mean(datos) #otra forma de calcular la media
    desviacion_media = sum(abs(x - media) for x in datos) / n
    return desviacion_media

# Generamos otro conjunto de datos aleatorios

conjunto_datos_otro = np.random.normal(loc=50, scale=10, size=100)

# Calculamos la media y la desviaciÃ³n media
media = np.mean(conjunto_datos_otro)
dm = desviacion_media(conjunto_datos_otro)

# Graficamos los datos
plt.figure(figsize=(10, 6))
plt.plot(conjunto_datos_otro, marker='o', linestyle='', label='Datos')
plt.axhline(y=media, color='r', linestyle='--', label='Media')
plt.axhline(y=media + dm, color='g', linestyle='--', label='Media + Desv. Media')
plt.axhline(y=media - dm, color='g', linestyle='--', label='Media - Desv. Media')
plt.xlabel('Ãndice del dato')
plt.ylabel('Valor')
plt.title('Datos junto con la Media y la DesviaciÃ³n Media')
plt.legend()
plt.grid(True)
plt.show()
```

>DesviaciÃ³n estÃ¡ndar

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

def calcular_varianza(datos):
    n = len(datos)
    media = np.mean(datos)
    varianza = sum((x - media)**2 for x in datos) / n
    return varianza

# Generamos un conjunto de datos aleatorios
# np.random.seed(2)
conjunto_datos = np.random.normal(loc=50, scale=10, size=100)

# Calculamos la media y la varianza
media = np.mean(conjunto_datos)
varianza = calcular_varianza(conjunto_datos)
desviacion_estandar = np.sqrt(varianza)

# Graficamos los datos
plt.figure(figsize=(10, 6))
plt.plot(conjunto_datos, marker='o', linestyle='', label='Datos')
plt.axhline(y=media, color='r', linestyle='--', label='Media')
plt.axhline(y=media + desviacion_estandar, color='g', linestyle='--', label='Media + Desv. EstÃ¡ndar')
plt.axhline(y=media - desviacion_estandar, color='g', linestyle='--', label='Media - Desv. EstÃ¡ndar')
plt.xlabel('Ãndice del dato')
plt.ylabel('Valor')
plt.title('Datos junto con la Media y la DesviaciÃ³n EstÃ¡ndar')
plt.legend()
plt.grid(True)
plt.show()

print("La varianza del conjunto de datos es:", varianza)
```

utilizar la desviaciÃ³n estandar y la desviaciÃ³n media puede darnos una idea de la dispercion de los datos y detectar outliders, dependiendo del caso usaremos uno u otro.

>Coeficiente de variaciÃ³n (CV)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar un conjunto de datos aleatorios
np.random.seed(0)
datos = np.random.normal(loc=50, scale=10, size=100)

# Calcular la media y la desviaciÃ³n estÃ¡ndar
media = np.mean(datos)
desviacion_estandar = np.std(datos)

# Calcular el coeficiente de variaciÃ³n (CV)
coeficiente_variacion = (desviacion_estandar / media) * 100

# Graficar los datos y marcar la media
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label='Media')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma con Media y Coeficiente de VariaciÃ³n')
plt.legend()
plt.grid(True)
plt.show()

print("Media:", media)
print("DesviaciÃ³n EstÃ¡ndar:", desviacion_estandar)
print("Coeficiente de VariaciÃ³n:", coeficiente_variacion)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# FunciÃ³n para calcular el coeficiente de variaciÃ³n
def coeficiente_variacion(datos):
    media = np.mean(datos)
    desviacion_estandar = np.std(datos)
    cv = (desviacion_estandar / media) * 100
    return cv

# Generar conjuntos de datos de ejemplo mÃ¡s grandes
np.random.seed(0)
datos1 = np.random.normal(loc=50, scale=10, size=1000)
datos2 = np.random.normal(loc=55, scale=5, size=1000)

# Calcular el coeficiente de variaciÃ³n para ambos conjuntos de datos
cv_datos1 = coeficiente_variacion(datos1)
cv_datos2 = coeficiente_variacion(datos2)

# Graficar histogramas para comparar los conjuntos de datos
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(datos1, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(datos1), color='red', linestyle='dashed', linewidth=1, label='Media')
plt.title('Histograma - Datos 1 (CV = {:.2f}%)'.format(cv_datos1))
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(datos2, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(datos2), color='red', linestyle='dashed', linewidth=1, label='Media')
plt.title('Histograma - Datos 2 (CV = {:.2f}%)'.format(cv_datos2))
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

# Aplicaciones de la EstadÃ­stica para la limpieza de datos

- IdentificaciÃ³n de *valores atÃ­picos* y errores de entrada: La estadÃ­stica descriptiva permite identificar valores atÃ­picos o errores de entrada mediante la exploraciÃ³n de medidas como la media, la mediana, la desviaciÃ³n estÃ¡ndar y los percentiles.

- ImputaciÃ³n de datos faltantes: La estadÃ­stica descriptiva proporciona mÃ©todos para manejar datos faltantes, como imputar valores utilizando tÃ©cnicas como la imputaciÃ³n media, la imputaciÃ³n mediana o la imputaciÃ³n de valores mÃ¡s frecuentes.
